```python
import numpy as np
import matplotlib.pyplot as plt

from dataLoader import importData
from shell_anon import NormalizedClusterLearner, NormalizedBayesClusterLearner
from normalization import InstanceNormalization, ErgoNormalization, NaiveNormalization, NoNormalization, PreTrainedNormalization
from evaluation import build_eval_set 


store_figures = '../stored_figures/normalization_ablation/'


data_set = 5

x_train, y_train, x_test, y_test, dataset_name = importData(data_set)
#x_train, x_test = pca_transform(x_train, x_test)

x_train = np.concatenate([x_train, x_test], axis=0)
y_train = np.concatenate([y_train, y_test])
num_class = np.max(y_train) + 1
print('Number of classes:', num_class)
```

```python
ind = 0
num_clus = 300
data, gt = build_eval_set(x_train, y_train, ind, 10)
```

```python
title_font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 12}

tic_font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 10}

plt.rc('font', **tic_font)
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
clf_noNorm = NormalizedBayesClusterLearner(num_clus = num_clus, norm = NoNormalization())
score = clf_noNorm.fit_and_display(data, gt)
```

```python
fig_name = store_figures + '/' + dataset_name + '_class_' + str(ind) + '_noNorm' +'.pdf'
clf_noNorm = NormalizedBayesClusterLearner(num_clus = num_clus, norm = NoNormalization())
score = clf_noNorm.fit_and_display(data, gt)
plt.savefig(fig_name,
            bbox_inches='tight',
            pad_inches = 0)
```

```python
fig_name = store_figures + '/' + dataset_name + '_class_' + str(ind) + '_NaiveNorm' +'.pdf'
clf_naiveNorm = NormalizedBayesClusterLearner(num_clus = num_clus, norm = NaiveNormalization())
score = clf_naiveNorm.fit_and_display(data, gt)

plt.savefig(fig_name,
            bbox_inches='tight',
            pad_inches = 0)
```

```python
fig_name = store_figures + '/' + dataset_name + '_class_' + str(ind) + '_InstanceNorm' +'.pdf'
clf_instanceNorm = NormalizedBayesClusterLearner(num_clus = num_clus, norm = InstanceNormalization())
score = clf_instanceNorm.fit_and_display(data, gt)

plt.savefig(fig_name,
            bbox_inches='tight',
            pad_inches = 0)
```

```python
fig_name = store_figures + '/' + dataset_name + '_class_' + str(ind) + '_ErgoNorm' +'.pdf'
clf_ergoNorm = NormalizedBayesClusterLearner(num_clus = num_clus, norm = ErgoNormalization())
score = clf_ergoNorm.fit_and_display(data, gt)

plt.savefig(fig_name,
            bbox_inches='tight',
            pad_inches = 0)
```
