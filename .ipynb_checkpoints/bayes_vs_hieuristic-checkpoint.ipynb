```python
import random
import pickle

import numpy as np
from dataLoader import importData
data_set = 0
store_folder = '../stored_stats/'
```

```python
x_train, y_train, x_test, y_test, dataset_name = importData(data_set)
#x_train, x_test = pca_transform(x_train, x_test)

x_train = np.concatenate([x_train, x_test], axis=0)
y_train = np.concatenate([y_train, y_test])
num_class = np.max(y_train) + 1
print('Number of classes:', num_class)
```

```python
from evaluation import build_eval_set


ind = 5
noise = 5
eval_data, eval_gt = build_eval_set(x_train, y_train, ind, noise)
```

```python
from normalization import InstanceNormalization
from shell_anon import NormalizedClusterLearner, NormalizedBayesClusterLearner
from sklearn.metrics import roc_auc_score

def eval_clus_impact(cluster_list, LearnerClass, eval_data, eval_gt):
    eval_results = []
    for num_clus in cluster_list:
        print('current number of clusters:', num_clus)
        clf = LearnerClass(num_clus = num_clus, norm = InstanceNormalization())
        clf.fit(eval_data)
        scores = clf.score_samples(eval_data)
        roc = roc_auc_score(eval_gt, scores)
        eval_results.append((num_clus, roc))
    return eval_results


cluster_list = [1, 2, 3, 4, 5, 10, 20, 100, 300, 500]   
eval_clus_results = eval_clus_impact(cluster_list, NormalizedClusterLearner, eval_data, eval_gt)
```

```python
cluster_list = [2, 3, 4, 5, 10, 20, 100, 300, 500]
eval_bayes_results = eval_clus_impact(cluster_list, NormalizedBayesClusterLearner, eval_data, eval_gt)
```

```python
import matplotlib.pyplot as plt

def plot_results(eval_results, axes=None, line_type = 'b-'):
    x_clus = []
    y_clus = []
    for r in eval_results:
        x_clus.append(r[0])
        y_clus.append(r[1])
     
    if axes is None:
        plt.plot(x_clus, y_clus, line_type)
    else:        
        axes.plot(x_clus, y_clus, line_type)



plot_results(eval_bayes_results, line_type = 'b-')  
plot_results(eval_clus_results, line_type = 'b--')  

plt.xscale('log')
# plt.yscale('log')
```

```python

with open(store_folder + dataset_name + '_' + str(ind) + '_' + str(noise) + '.pickle', 'wb') as f:
    pickle.dump((eval_clus_results,eval_bayes_results), f)

```

```python
file1 = '../stored_stats/mnist_5.pickle'
file2 = '../stored_stats/STL-10_1.pickle'
file3 = '../stored_stats/fashion-mnist_5_5.pickle'
file4 = '../stored_stats/MIT-Places-Small_3.pickle'

with open(file1, 'rb') as f: 
    (eval_clus_mnist, eval_bayes_mnist) = pickle.load(f)

with open(file2, 'rb') as f: 
    (eval_clus_stl, eval_bayes_stl) = pickle.load(f)
    
with open(file3, 'rb') as f: 
    (eval_clus_fmnist, eval_bayes_fmnist) = pickle.load(f)
    
with open(file4, 'rb') as f: 
    (eval_clus_mit, eval_bayes_mit) = pickle.load(f)
```

```python
title_font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 12}

tic_font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 10}

plt.rc('font', **tic_font)




figure, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
axes[0].set_title('Dist. to Cluster Mean')
plot_results(eval_clus_mnist, axes[0],'k--')  
plot_results(eval_clus_fmnist, axes[0], 'k:')  
#plot_results(eval_clus_stl, axes[0], 'k-.')  
plot_results(eval_clus_mit, axes[0], 'k-.')  

axes[0].set_ylabel('AUROC')
axes[0].set_xlabel('Log Number of Clusters')

axes[1].set_title('LVAD')
plot_results(eval_bayes_mnist, axes[1],'k--')  
plot_results(eval_bayes_fmnist, axes[1], 'k:')  
#plot_results(eval_bayes_stl, axes[1], 'k-.')  
plot_results(eval_bayes_mit, axes[1], 'k-.')  

axes[1].yaxis.tick_right()
axes[1].axis('on')
axes[1].set_xlabel('Log Number of Clusters')


axes[1].legend(['Mnist', 'F-MNIST', 'MIT-Places'], loc = 'lower right')

plt.xscale('log')

#plt.yscale('log')
plt.tight_layout() 
plt.figure(figsize=(50, 200))
plt.savefig('../stored_figures/clustering_problem/mnist_fmnist_mit.pdf')
```
