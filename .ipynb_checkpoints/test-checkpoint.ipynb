```python
from tf_features import generator2feature, folders2generator, download_traditional_network
import numpy as np

im_dir = '../../fake-stl10/'
#im_dir = '../../MIT-Places-Small//'

gen = folders2generator(im_dir, batch_size = 640, target_class=[0])
```

```python
model = download_traditional_network()
feats, gts, im_paths = generator2feature(model, gen)
```

```python
feats = np.concatenate(feats, axis=0)
gts = np.concatenate(gts)
im_paths = sum(im_paths, [])
```

```python
len(im_paths)
```

```python
feats.shape
```

```python
from normalization import NormalizedAnomalyDetector
from shell_anomaly_detectors import LVAD

nlvad = NormalizedAnomalyDetector(clf = LVAD())
nlvad.fit(feats)
```

```python
scores = nlvad.score_samples(feats)
```

```python
ranks = np.argsort(scores)
```

```python
from PIL import Image

full_paths = [im_dir + '/' + im_path for im_path in im_paths]
```
