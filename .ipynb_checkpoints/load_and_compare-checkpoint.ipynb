```python
import numpy as np
import pickle
from dataLoader import importData
from evaluation import AnonEvaluationStatistics
import matplotlib.pyplot as plt

store_folder = '../stored_results/'
store_figures = '../stored_figures/compare_dist_based_algo/'
```

```python
def loader(algo_name, dataset_name, store_folder):
    store_path = store_folder + '/' + dataset_name + algo_name + '.pickle'
    with open(store_path, 'rb') as file:
        anon_eval = pickle.load(file)
    return anon_eval

algo_names = ['_bayes_instance_normalization_300', 
              '_cluster_affinity_instance_normalization_', 
              '_affinity_instance_normalization_',
              '_isolationforest_normalization_',
              '_ocsvm',
              '_DAGMM_normalization_',
              '_LOF_normalization_', 
              '_RSRAE_instanceNorm']


data_set = 1
_, _, _, _, dataset_name = importData(data_set)
```

```python
anon_eval_bayes = loader(algo_names[0], dataset_name, store_folder)
anon_eval_naive_cluster = loader(algo_names[1], dataset_name, store_folder)
anon_eval_naive_euclidean = loader(algo_names[2], dataset_name, store_folder)
anon_eval_isoforest = loader(algo_names[3], dataset_name, store_folder)
anon_eval_ocsvm = loader(algo_names[4], dataset_name, store_folder)
anon_eval_dagmm = loader(algo_names[5], dataset_name, store_folder)
anon_eval_lof = loader(algo_names[6], dataset_name, store_folder)
anon_eval_rsrae = loader(algo_names[7], dataset_name, store_folder)
```

```python
title_font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 12}

tic_font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 10}
```

```python
def plt_error_bars(anon_eval, axes=None, linestyle='', title='', xlabel='', ylabel=''):
    top_err = np.max(anon_eval.auroc, axis=0) - np.mean(anon_eval.auroc, axis=0)
    bot_err = np.mean(anon_eval.auroc, axis=0) - np.min(anon_eval.auroc, axis=0)
    err = np.array([bot_err, top_err] ) /10
    if axes is None:
        plt.errorbar(anon_eval.percentiles, np.mean(anon_eval.auroc, axis=0), err,
                     fmt = linestyle, capsize=5, capthick=True)
    else:
        axes.errorbar(anon_eval.percentiles, 
                      np.mean(anon_eval.auroc, axis=0), 
                      err, 
                      fmt = linestyle, capsize=5, capthick=True)
        axes.set_title(title)
        axes.set_xlabel(xlabel)
        axes.set_ylabel(ylabel)

        
plt.rc('font', **tic_font)

figure, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
plt_error_bars(anon_eval_isoforest, axes[0], '--r', 'Isolation \n Forest')
plt_error_bars(anon_eval_bayes, axes[1], '--b', 'Bayesian \n Inference')
plt.xscale('log')

figure, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
plt_error_bars(anon_eval_naive_euclidean, axes[0], '--r', 'Naive \n Euclidean')
plt_error_bars(anon_eval_bayes, axes[1], '--b', 'Bayesian \n Inference')
plt.xscale('log')


figure, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
plt_error_bars(anon_eval_naive_cluster, axes[0], '--r', 'Naive \nCluster')
plt_error_bars(anon_eval_bayes, axes[1], '--b', 'Bayesian \n Inference')
plt.xscale('log')

#plt.savefig('Bayes_vs Naive_Euc' + dataset_name + '.pdf' ,bbox_inches='tight',pad_inches = 0)

```

```python

plt.rc('font', **tic_font)
figure, axes = plt.subplots(nrows=1, ncols=5, sharex=True, sharey=True)

plt_error_bars(anon_eval_isoforest, axes[0], '--r', 'Isolation \n Forest', ylabel='Average AUROC')
plt_error_bars(anon_eval_ocsvm, axes[1], '--y', 'One-Class \n SVM')
plt_error_bars(anon_eval_dagmm, axes[2], '--g', 'DAGMM \n ', xlabel='Anomaly Percentage')
plt_error_bars(anon_eval_rsrae, axes[3], '--m', 'RSRAE \n')
plt_error_bars(anon_eval_bayes, axes[4], '--b', 'LVAD \n (Ours)')
plt.xscale('log')


plt.savefig(store_figures + 'Bayes vs DAGMM vs OCSVM vs RSRAE Iso_' + dataset_name + '.pdf',
            bbox_inches='tight',
            pad_inches = 0)
```
